```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
gas = pd.read_csv('gas_price/gas_prices.csv')
```

```python
plt.title('Valores do combust√≠vel em US dollar')

plt.plot(gas['Year'], gas['Australia'], 'r.-', label = 'Australia')
plt.plot(gas['Year'], gas['Italy'], label = 'Italy')
plt.plot(gas['Year'], gas['USA'], label = 'USA')

plt.xlabel('Years')
plt.ylabel('US dollar')

plt.xticks(gas['Year'][::2])  #do inicio ao fim de 2 em 2 anos
plt.legend()
plt.show()
```

```python
plt.title('Valores do combust√≠vel em US dollar')

list_country = ['Australia', 'USA', 'Canada']

for country in gas:
    if country in list_country:
        plt.plot(gas['Year'], gas[country], label = [country], marker = '.')

plt.xlabel('Years')
plt.ylabel('US dollar')

plt.xticks(gas['Year'][::2])  #do inicio ao fim de 2 em 2 anos
plt.legend()
plt.savefig('country.png')
plt.show()
```
